### Input Normalization
-    255     0~1  .

```python
def normalize_and_resize_img(image, label):
    """Normalizes images: `uint8` -> `float32`."""
    # image = tf.image.resize(image, [32, 32])
    return tf.cast(image, tf.float32) / 255., label
```

```python
def apply_normalize_on_dataset(ds, is_test=False, batch_size=16):
    ds = ds.map(
        normalize_and_resize_img, 
        num_parallel_calls=1
    )
    ds = ds.batch(batch_size)
    if not is_test:
        ds = ds.repeat()
        ds = ds.shuffle(200)
    ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
    return ds
```

```python
# block_num    , input_shape summary   .
# VGG block

def build_vgg_block(input_layer,
                   num_cnn=3,
                   channel=64,
                   block_num=1):
    # input layer
    x = input_layer
    
    # CNN layer
    for cnn_num in range(num_cnn):
        x = keras.layers.Conv2D(filters=channel,
                               kernel_size=(3,3),
                               activation='relu',
                               kernel_initializer='he_normal',
                               padding='same',
                               name=f'block{block_num}_conv{cnn_num}'
                               )(x)
        
    # max pooling layer
    x = keras.layers.MaxPooling2D(pool_size=(2,2),
                                 strides=2,
                                 name=f'block{block_num}_pooling'
                                 )(x)
    return x
```

```python
vgg_input_layer = keras.layers.Input(shape=(32,32,3))
vgg_block_output = build_vgg_block(vgg_input_layer)
```

```python
model = keras.Model(inputs=vgg_input_layer, outputs = vgg_block_output)

model.summary()
```

## Ablation Study  3. VGG Complete Model

### 1. VGG-16

-  CNN      .

```python
# VGG    
def build_vgg(input_shape=(32,32,3),
             num_cnn_list=[2,2,3,3,3],
             channel_list=[64,128,256,512,512],
             num_classes=10):
    
    assert len(num_cnn_list) == len(channel_list)
    input_layer = keras.layers.Input(shape=input_shape)
    output = input_layer
    
    for i, (num_cnn, channel) in enumerate(zip(num_cnn_list, channel_list)):
        output = build_vgg_block(output,
                                num_cnn=num_cnn,
                                channel=channel,
                                block_num=i)
        
        output = keras.layers.Flatten(name='flatten')(output)
        output = keras.layers.Dense(4096, activation='relu', name='fc1')(output)
        output = keras.layers.Dense(4096, activation='relu', name='fc2')(output)
        output = keras.layers.Dense(num_classes, activation='softmax', name='predictions')(output)
        
        model = keras.Model(inputs=input_layer, outputs=output)
        
        return model
```

```python
vgg_16 = build_vgg()
vgg_16.summary()
```

### VGG-19
-   VGG-19 .

```python
vgg_19 = build_vgg(num_cnn_list=[2,2,4,4,4],
                  channel_list=[64,128,256,512,512])

vgg_19.summary()
```

# Ablation Study  4. VGG16 vs VGG19

- CIFAR 10  .

```python
BATCH_SIZE=256
EPOCH = 20
```

```python
# trainset, testset
(ds_train, ds_test), ds_info = tfds.load('cifar10',
                                        split=['train','test'],
                                        as_supervised=True,
                                        shuffle_files=True,
                                        with_info=True)

ds_train = apply_normalize_on_dataset(ds_train, batch_size=BATCH_SIZE)
ds_test = apply_normalize_on_dataset(ds_test, batch_size=BATCH_SIZE)
```

```python
vgg_16.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=tf.keras.optimizers.SGD(lr=0.01, clipnorm=1.),
    metrics=['accuracy'],
)

history_16 = vgg_16.fit(
    ds_train,
    steps_per_epoch=int(ds_info.splits['train'].num_examples/BATCH_SIZE),
    validation_steps=int(ds_info.splits['test'].num_examples/BATCH_SIZE),
    epochs=EPOCH,
    validation_data=ds_test,
    verbose=1,
    use_multiprocessing=True,
)
```

```python
vgg_19.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=tf.keras.optimizers.SGD(lr=0.01, clipnorm=1.),
    metrics=['accuracy'],
)

history_19 = vgg_19.fit(
    ds_train,
    steps_per_epoch=int(ds_info.splits['train'].num_examples/BATCH_SIZE),
    validation_steps=int(ds_info.splits['test'].num_examples/BATCH_SIZE),
    epochs=EPOCH,
    validation_data=ds_test,
    verbose=1,
    use_multiprocessing=True,
)
```

### VGG-16 19 

```python
# training loss 

import matplotlib.pyplot as plt
plt.plot(history_16.history['loss'], 'r')
plt.plot(history_19.history['loss'], 'b')
plt.title('Model training loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['vgg_16','vgg_19'], loc='upper left')
plt.show()
```

```python
# validation acc 

plt.plot(history_16.history['val_accuracy'], 'r')
plt.plot(history_19.history['val_accuracy'], 'b')
plt.title('model valid acc')
plt.ylabel('Acc')
plt.xlabel('epoch')
plt.legend(['vgg_16','vgg_19'], loc='upper left')
plt.show()
```
